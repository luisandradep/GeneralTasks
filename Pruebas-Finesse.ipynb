```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

```python
l2 = pd.read_excel("Prueba-empaques-Finesse.xlsx", sheet_name='L2', header=0)
l2['Fecha'] = pd.to_datetime(l2['Fecha'])
l2
```

```python
l2.dtypes
```

```python
mask1 = l2['Moho']==0
mask2 = l2['Moho']==1
```

```python
fig, ax = plt.subplots(figsize=(12,7))
ax.bar(l2['Fecha'][mask1], l2['Cantidad'][mask1], width=0.5, color= 'blue', label='Sin moho')
ax.bar(l2['Fecha'][mask2], l2['Cantidad'][mask2], width=0.5, color= 'red', label='Con Moho')
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m'))
ax.set_xlabel('Fecha de consumo')
ax.set_ylabel('Cantidad de tajadas consumidas')
plt.title('Cantidad de tajadas consumidas de empaque L2 por fecha y condici√≥n de tajada')
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```
